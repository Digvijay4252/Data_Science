```python
import pandas as pd
```

```python
df=pd.read_csv('/content/Student_Performance.csv')
df.head()
```

```python
df.info()
```

```python
df.notnull().sum()
```

```python
import matplotlib.pyplot as plt

plt.bar(df['Hours Studied'], df['Performance Index'])
plt.ylabel('Hours Studied')
plt.xlabel('Performance Index')
plt.title('Hours Studied vs. Performance Index')
plt.show()
```

```python
from sklearn.linear_model import LinearRegression
from sklearn.model_selection import train_test_split

# df['Extracurricular Activities']=df['Extracurricular Activities'].map({'Yes':1,'No':0})

# df.notnull().sum()
x=df.drop(['Performance Index','Extracurricular Activities'],axis=1)
y=df['Performance Index']
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.2)

model = LinearRegression()
model.fit(x_train, y_train)

predictions = model.predict(x_test)
print("Logistic Regression predictions: Performance Index: ", predictions[:5])
```

```python
import numpy as np
from sklearn.metrics import r2_score, mean_squared_error

predictions = model.predict(x_test)

r2 = r2_score(y_test, predictions)
print(f"R-squared: {r2}")

rmse = np.sqrt(mean_squared_error(y_test, predictions))
print(f"RMSE: {rmse}")
```

```python
from sklearn.tree import DecisionTreeRegressor

x=df.drop(['Performance Index','Extracurricular Activities'],axis=1)
y=df['Performance Index']
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.2)

model = DecisionTreeRegressor()
model.fit(x_train, y_train)

predictions = model.predict(x_test)
print("Decision Tree predictions: Performance Index: ", predictions[:5])
```

```python
predictions = model.predict(x_test)

r2 = r2_score(y_test, predictions)
print(f"R-squared: {r2}")

rmse = np.sqrt(mean_squared_error(y_test, predictions))
print(f"RMSE: {rmse}")
```

```python
from sklearn.ensemble import RandomForestRegressor

x=df.drop(['Performance Index','Extracurricular Activities'],axis=1)
y=df['Performance Index']
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.2)

model = RandomForestRegressor()
model.fit(x_train, y_train)

predictions = model.predict(x_test)
print("Random Forest Regressor predictions: Performance Index: ", predictions[:5])
```

```python
predictions = model.predict(x_test)

r2 = r2_score(y_test, predictions)
print(f"R-squared: {r2}")

rmse = np.sqrt(mean_squared_error(y_test, predictions))
print(f"RMSE: {rmse}")
```
